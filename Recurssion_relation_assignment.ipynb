```python
#3. Given a recurrence relation, solve it using the recursive tree approach:

a. T(n) = 2T(n-1) +1
first,Draw the recursive tree for the recurrence relation 
                     
                      T(n)
                     /   \
                 T(n-1)   T(n-1)
                   / \     / \
               T(n-2) T(n-2) T(n-2) T(n-2)
                / \    / \    / \    / \
           T(n-3) ... T(n-3) T(n-3) T(n-3)
At each Level The Recurrence branches into two subproblems of  size n - 1,and the work done at each level is 1.
The total number of levels in recursive tree is n. Lets count the work done at each level.
Level 0 = 1 work
Level 1 = 2 work
Level 2 = 2^2 work
Level 3 = 2^3 work
.....
Level k = 2^k work
The total work done is the sum of these geometric series
1 + 2 + 2^2 + 2^3 +....+ 2^(n-1) = 2^n -1
The solution is T(n) = 2^n - 1

b. T(n) = 2T(n/2) + n

                   T(n)
                   / \
             2T(n/2) 2T(n/2)
              / \      / \
          T(n/4) ... T(n/4) T(n/4) ...

At each level, the recurrence branches into two subproblems of size n/2 and the work done by each level is n.
The total number of recursion tree level is log(n) base 2.now counts the work done at each level.
Level 0 = n work
Level 1 = 2.n/2 = n work
Level 2 = 2^2.n/2^2 = n works
Level 3 = 2^3.n/2^3 = n works
..............
Level k = 2^k.n/2^k = n works
The total work done is the product of the work at each level and the number of nodes at each level:
n.log(n) Base 2

Therefore , the solution is T(n) = n.log base2(n)
    
```
